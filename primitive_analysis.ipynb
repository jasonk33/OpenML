```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%load_ext autotime
%matplotlib inline  
```

```python
"""
Workflow:
- Boxplot of sklearn primitive accuracies
- Sklearn primitive accuracies dataframe
- Boxplot of weka primitive accuracies
- Weka primitive accuracies dataframe
"""
```

```python
# Read in sklearn flow accuracies
flow_accuracies_df = pd.read_csv("OpenML_Data/sklearn_flow_accuracies.csv")
mean_flow_accuracies = pd.read_csv("OpenML_Data/adjusted_flow_accuracies.csv")
```

```python
# Data manipulation for plotting
counts = mean_flow_accuracies.sort_values(by=['Adj_Accuracy'], ascending=False)['Count']
names = mean_flow_accuracies.sort_values(by=['Adj_Accuracy'], ascending=False)['Name']
names_counts = []
for idx in range(len(counts)):
    names_counts.append(names[idx] + " : " + str(counts[idx]))
```

```python
# Boxplots for adjusted accuracy of sklearn flows
sns.set(style="ticks")
f, ax = plt.subplots(figsize=(25, 25))
plot = sns.boxplot(y="Name", x="Adj_Accuracy", data = flow_accuracies_df, order=mean_flow_accuracies['Name'])
plot.set_xlabel("Accuracy",fontsize=30)
plot.set_ylabel("Primitive",fontsize=30)
plot.set_title("Sklearn Primitives vs Adjusted Accuracy", fontsize=50)
plot.tick_params(labelsize=20)
plot.set_yticklabels(names_counts)
plot.plot([0,0], [-5,35], 'r--', linewidth = 5)
```

```python
mean_flow_accuracies.sort_values(by=['Adj_Accuracy'], ascending=False).iloc[:,1:]
```

```python
# Read in weka flow accuracies
flow_accuracies_df = pd.read_csv("OpenML_Data/weka_flow_accuracies_df.csv")
mean_flow_accuracies = pd.read_csv("OpenML_Data/weka_mean_flow_accuracies.csv")
```

```python
# Data manipulation for plotting
counts = mean_flow_accuracies.sort_values(by=['Adj_Accuracy'], ascending=False)['Count']
names = mean_flow_accuracies.sort_values(by=['Adj_Accuracy'], ascending=False)['Name']
names_counts = []
for idx in range(len(counts)):
    names_counts.append(names[idx] + " : " + str(counts[idx]))
```

```python
# Boxplots for adjusted accuracy of weka flows
sns.set(style="ticks")
f, ax = plt.subplots(figsize=(35, 35))
plot = sns.boxplot(y=flow_accuracies_df["Name"], x=flow_accuracies_df["Adj_Accuracy"], order=names)
plot.set_xlabel("Accuracy",fontsize=30)
plot.set_ylabel("Primitive",fontsize=40)
plot.set_title("Weka Primitives vs Adjusted Accuracy", fontsize=50)
plot.tick_params(labelsize=25)
plot.set_yticklabels(names_counts)
plot.plot([0,0], [-25,55], 'r--', linewidth = 5)
```
