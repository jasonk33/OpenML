```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
from scipy.stats import kurtosis, skew, entropy
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
from sklearn.preprocessing import scale
import os
%load_ext autotime
%matplotlib inline 
```

```python
"""
Workflow:
- Fit kmeans models on all the dataset
- Select the number of clusters to be used in kmeans model
- Use the dataset cluster to structure weka data by cluster
- Plot the adjusted weka flow accuracies by dataset cluster
"""
```

```python
# Load dataset features
df_data_sets = pd.read_csv("OpenML_Data/paper_data_set_features.csv", index_col=0)
```

```python
# Cluster the datasets using kmeans on their features

# Max clusters to try
components = 20

#convert it to numpy arrays
X=df_data_sets.values

#Scaling the values
X = scale(X)

distortions = []
K = range(1,components)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k for k means')
plt.show()
```

```python
# Fit the kmeans model to the data with the selected number of clusters
kmeanModel = KMeans(n_clusters=8).fit(X)
kmeanModel.fit(X)
```

```python
# Store the cluster labels for each dataset
data_set_clusters = {}
for idx,label in enumerate(kmeanModel.labels_):
    data_set_clusters[str(df_data_sets.iloc[idx,:].name)] = int(label)
    
# with open('OpenML_Data/data_set_clusters.json', 'w') as outfile:
#     json.dump(data_set_clusters, outfile)
```

```python
### Below is code that was run on Brown CCV environment to format weka data using the dataset clusters
```

```python
# Load the necessary data
with open('weka_run_descriptions.json') as data_file:
    weka_run_descriptions = json.load(data_file)
    
with open('sklearn_task_descriptions.json') as data_file:
    sklearn_task_descriptions = json.load(data_file)
    
with open('data_set_clusters.json') as data_file:
    data_set_clusters = json.load(data_file)
    
with open('weka_flow_descriptions.json') as data_file:
    weka_flow_descriptions = json.load(data_file)
```

```python
# Get the runs for each dataset cluster
weka_run_descriptions_clustered = {}
for i in range(0,8):
    weka_run_descriptions_clustered[str(i)] = {}
for run_id,run in weka_run_descriptions.items():
    task_id = run['run']['task_id']
    data_set_id = int(sklearn_task_descriptions[task_id]['task']['input'][0]['data_set']['data_set_id'])
    try:
        weka_run_descriptions_clustered[str(data_set_clusters[str(data_set_id)])][run_id] = run
    except:
        pass

# with open('weka_run_descriptions_clustered.json', 'w') as outfile:
#     json.dump(weka_run_descriptions_clustered, outfile)
```

```python
# Load the accuracies of all runs for each task
task_accuracies = {}
for file in os.listdir("results"):
    with open('results/' + file) as data_file:
        acc = json.load(data_file)
    task_accuracies.update(acc)
```

```python
# For each cluster, get the adjusted weka flow accuracies
for cluster_label in range(0,8):
    
    # Get all weka primitives for each run
    flows_per_run = {}
    for run_id,run_description in weka_run_descriptions_clustered[str(cluster_label)].items():
        flow_id = run_description['run']['flow_id']
        flow_names = []
        flow_description = weka_flow_descriptions[flow_id]['flow']
        name = flow_description['name']
        flow_names.append(name)
        flows_per_run[run_id] = flow_names

    # Get the number of times each weka flow is run
    flow_names = {}
    for run,flows in flows_per_run.items():
        for flow in flows:
            if flow in flow_names:
                flow_names[flow] += 1
            else:
                flow_names[flow] = 1
    
    # Get the adjusted average accuracy (flow accuracy minus mean task accuracy) for all weka flows
    weka_flow_accuracies = {}
    for run_id, run_description in weka_run_descriptions_clustered[str(cluster_label)].items():
        for metric in run_description['run']['output_data']['evaluation']:
            if metric['name'] == 'predictive_accuracy':
                accuracy = float(metric['value'])
                try:
                    avg_task_acc = task_accuracies[run_description['run']['task_id']]
                    adjusted_accuracy = (accuracy - avg_task_acc)/avg_task_acc
                    for flow in flows_per_run[run_id]:
                        if flow in weka_flow_accuracies:
                            weka_flow_accuracies[flow].append(adjusted_accuracy)
                        else:
                            weka_flow_accuracies[flow] = [adjusted_accuracy]
                    break
                except:
                    pass
    
    # Create dataframe of average weka flow adjusted accuracies
    mean_flow_accuracies = pd.DataFrame(columns=['Name', 'Adj_Accuracy', 'Count', 'sd'])
    for flow,accuracies in weka_flow_accuracies.items():
        mean_flow_accuracies = mean_flow_accuracies.append(pd.DataFrame([[flow,np.asarray(accuracies).astype(np.float).mean(), len(accuracies), np.asarray(accuracies).astype(np.float).std()]], columns=['Name', 'Adj_Accuracy', 'Count', 'sd']))
    mean_flow_accuracies = mean_flow_accuracies[mean_flow_accuracies['Count'] >= 0]

    mean_flow_accuracies.to_csv('OpenML_Data/clustered_accuracies/weka_mean_flow_accuracies_' + str(cluster_label) + '.csv')

    # Reformat data
    flow_accuracies_df = pd.DataFrame(columns=['Name', 'Adj_Accuracy'])
    for flow,accuracies in weka_flow_accuracies.items():
        for accuracy in accuracies:
            flow_accuracies_df = flow_accuracies_df.append(pd.DataFrame([[flow,accuracy]], columns=['Name', 'Adj_Accuracy']))
    flow_accuracies_df = flow_accuracies_df[flow_accuracies_df.groupby('Name').Adj_Accuracy.transform(len) > 0]

    flow_accuracies_df.to_csv('OpenML_Data/clustered_accuracies/weka_flow_accuracies_df' + str(cluster_label) + '.csv')
```

```python
# Plot the adjusted weka flow accuracies for each dataset cluster
for label in range(0,8):
    mean_flow_accuracies = pd.read_csv("OpenML_Data/clustered_accuracies/weka_mean_flow_accuracies_{}.csv".format(label))
    mean_flow_accuracies = mean_flow_accuracies.sort_values('Count', ascending=False).iloc[0:25,:]
    flow_accuracies_df = pd.read_csv("OpenML_Data/clustered_accuracies/weka_flow_accuracies_df{}.csv".format(label))
    flow_accuracies_df = flow_accuracies_df[flow_accuracies_df['Name'].isin(mean_flow_accuracies.Name)]
    
    counts = mean_flow_accuracies.sort_values(by=['Adj_Accuracy'], ascending=False)['Count']
    names = mean_flow_accuracies.sort_values(by=['Adj_Accuracy'], ascending=False)['Name']
    names_counts = []
    for idx in range(len(counts)):
        names_counts.append(list(names)[idx] + " : " + str(list(counts)[idx]))

    sns.set(style="ticks")
    f, ax = plt.subplots(figsize=(35, 35))
    plot = sns.boxplot(y=flow_accuracies_df["Name"], x=flow_accuracies_df["Adj_Accuracy"], order=names)
    plot.set_xlabel("Accuracy",fontsize=30)
    plot.set_ylabel("Primitive",fontsize=40)
    plot.set_title("Weka Primitives vs Adjusted Accuracy (Cluster {})".format(label), fontsize=50)
    plot.tick_params(labelsize=25)
    plot.set_yticklabels(names_counts)
    plot.plot([0,0], [-25,55], 'r--', linewidth = 5)
```
