```python
import numpy as np
import pandas as pd
```

```python
#Supervised Classification on credit-g
#https://www.openml.org/t/31
```

```python
df = pd.read_csv("Results/OpenML Supervised Classification on credit-g.csv")
df.groupby(['Flow'])['Flow','predictive_accuracy'].mean().sort_values(by=['predictive_accuracy'], ascending=False)
```
